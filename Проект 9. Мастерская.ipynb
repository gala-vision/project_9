Функция для вывода основной информации о датафрейме:

Функция для построения горизонтальной столбчатой диаграммы, которая покажет количество пропусков в столбцах датафрейма в процентном соотношении:

```python
def missing_data_plot(data, color):
    # рассчитываем % пропущенных значений для каждого столбца
    missing_data = (data.isna().sum() / len(data) * 100).sort_values()
    plt.figure(figsize=(15, 10))
    plt.title('Процент пропущенных значений по столбцам')
    plt.xlabel('количество пропусков, %')
    plt.ylabel('название столбца')
    plt.barh(missing_data.index, missing_data, color=color)

    plt.show()
```

Функция для поиска похожих значений с помощью расстояния Левенштейна поможет искать неявные дубликаты в данных:

```python
def find_similar_values(column, max_distance=1):
    unique_values = df_train[column].unique()
    similar_values = []

    for i, value in enumerate(unique_values):
        # Пропускаем, если значение пропущено
        if pd.isna(value):
            continue
        
        for other_value in unique_values[i+1:]:
            # Пропускаем, если другое значение пропущено
            if pd.isna(other_value):
                continue
            
            dist = distance(value, other_value)
            if dist <= max_distance:
                similar_values.append((value, other_value, dist))
    
    return similar_values
```

Функция для построения гистограммы и диаграммы размаха для проведения исследовательского анализа количественных признаков:

```python
def hist_boxplot(data, features, color='lightseagreen'):
    for col in features:
        fig, axs = plt.subplots(1, 2, figsize=(20, 5))

        # Гистограмма
        axs[0].hist(data[col], color=color, edgecolor='black', alpha=0.5, range=(2, data[col].max()))
        axs[0].set_title(f'Гистограмма для {col}', fontsize=15)
        axs[0].set_xlabel(col)  
        axs[0].set_ylabel('count')  

        # Диаграмма размаха
        sns.boxplot(x=data[col], ax=axs[1], color=color)
        axs[1].set_title(f'Диаграмма размаха для {col}', fontsize=15)
        axs[1].set_xlabel(col)  

        plt.show()
```

Функция для удаления выбросов:

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

# Обучение модели

```python
# Препроцессор
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value=-1)),
    ('scaler', StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, num_columns)
], remainder='passthrough')

# Разделение данных
X_train = df_train_model.drop(['name', 'status'], axis=1)
y_train = df_train_model['status']
X_test = df_test_model.drop(['name'], axis=1)
```

```python
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
```

Применим SMOTE для оверсэмплинга

```python
smote = SMOTE(random_state=RANDOM_STATE)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_preprocessed, y_train)
```

Будем использовать модель градиентного бустинга. С помощью случайного поиска найдем для нее в пайплайне лучшие параметры:

```python
models = {
    'GradientBoosting': GradientBoostingClassifier()
}

param_distributions = {
    'GradientBoosting': {
        'classifier__n_estimators': randint(100, 120),
        'classifier__learning_rate': uniform(0.01, 0.1),
        'classifier__max_depth': [3, 5, 7]
    }
}
```

```python
# Определение метрики F1
f1 = make_scorer(f1_score, average='binary', pos_label='closed')
```

```python
best_models = {}
for model_name in models:
    pipeline = Pipeline(steps=[
        ('classifier', models[model_name])
    ])
    random_search = RandomizedSearchCV(pipeline, param_distributions[model_name], scoring=f1, cv=5, n_jobs=-1, n_iter=10, random_state=RANDOM_STATE)
    random_search.fit(X_train_resampled, y_train_resampled)
    best_models[model_name] = random_search.best_estimator_
```

```python
random_search.best_params_
```

```python
startup_names = df_test_model['name']
```

```python
for model_name in best_models:
    y_pred = best_models[model_name].predict(X_test_preprocessed)
    precision = precision_score(y_train_resampled, best_models[model_name].predict(X_train_resampled), pos_label='closed')
    recall = recall_score(y_train_resampled, best_models[model_name].predict(X_train_resampled), pos_label='closed')
    f1 = f1_score(y_train_resampled, best_models[model_name].predict(X_train_resampled), pos_label='closed')
    print('Метрики для модели', model_name, 'Precision:', round(precision, 2), 'Recall:', round(recall, 2), 'f1 Score:', round(f1, 2))

    predictions_df = pd.DataFrame({
        'name': startup_names,
        'status': y_pred
    })

    # Сохранение предсказаний для загрузки на платформу Kaggle
    predictions_df.to_csv(f'{model_name}_predictions.csv', index=False)
```
